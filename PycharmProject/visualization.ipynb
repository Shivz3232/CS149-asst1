```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Mandlebrot Threads

```python
# Load benchmark results
view1 = pd.read_csv("../prog1_mandelbrot_threads/benchmark_view_1.csv")
view2 = pd.read_csv("../prog1_mandelbrot_threads/benchmark_view_1.csv")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# View 1
ax1.plot(view1["threads"], view1["speedup"], marker="o")
ax1.plot(view1["threads"], view1["threads"], "--", label="Ideal")
ax1.set_title("View 1")
ax1.set_xlabel("Threads")
ax1.set_ylabel("Speedup")
ax1.grid(True)
ax1.legend()

# View 2
ax2.plot(view2["threads"], view2["speedup"], marker="o")
ax2.plot(view2["threads"], view2["threads"], "--", label="Ideal")
ax2.set_title("View 2")
ax2.set_xlabel("Threads")
ax2.set_ylabel("Speedup")
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```
